```python
#Visualize the Data

import matplotlib.pyplot as plt
import seaborn as sns

correlation = df['Total Household Income'].corr(df['Total Food Expenditure'])
print(f"Correlation between income and consumption: {correlation}")

# Check for zero or infinite values
#there are 60 rows where the Total Household Income is less than or equal to 0
#there are 60 rows where the Total Food Expenditure is less than or equal to 0.
#print(df[df['Total Household Income'] <= 0].shape)
#print(df[df['Total Food Expenditure'] <= 0].shape)

# Replace 0 or negative values in 'Total Household Income' and 'Total Food Expenditure' with NaN
#df['Total Household Income'] = df['Total Household Income'].replace(0, np.nan)
#df['Total Food Expenditure'] = df['Total Food Expenditure'].replace(0, np.nan)

# Drop rows with NaN values
df_cleaned = df.dropna(subset=['Total Household Income', 'Total Food Expenditure'])


#Scatter plot: Income vs. Consumption
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Total Household Income'], y=df['Total Food Expenditure'])
plt.xlabel("Income")
plt.ylabel("Food Consumption")
plt.title("Income vs. Food Consumption")
plt.show()
```

```python
#Run Regression Analysis

import statsmodels.api as sm

#Define independent (X) and dependent (y) variables
X = df["Total Household Income"]
y= df["Total Food Expenditure"]

#Add a constant
X = sm.add_constant(X)

#Rund OLS regression
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Log-transform the variables
df['Log Total Household Income'] = np.log(df['Total Household Income'] + 1)
df['Log Total Food Expenditure'] = np.log(df['Total Food Expenditure'] + 1)

# Run the regression again
X_log = sm.add_constant(df['Log Total Household Income'])
y_log = df['Log Total Food Expenditure']
model_log = sm.OLS(y_log, X_log).fit()

print(model_log.summary())
```

```python
#- The log-transformed model is a better fit, explaining more of the variance (66%) in food expenditure compared to the original model (44%).
#- There is a strong, positive relationship between income and food expenditure, with the log transformation making the model more interpretable in terms of percentage changes.
#- The model is statistically significant, but the residuals might need further analysis for normality, especially if you're using this model for hypothesis testing.
```
